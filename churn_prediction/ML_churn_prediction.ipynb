```python
import pandas as pd
```

```python
contracts = pd.read_csv("ContractData.csv", sep = ",")
```

```python
contracts
```

```python
calls = pd.read_csv("CallsData.csv", sep = ";", decimal=',')
```

```python
calls
```

```python
# juntar os datasets
dataset = pd.merge(contracts, calls, how = 'inner', on = ['Phone', 'Area Code'])
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
dataset
```

```python
dataset['Charge_per_day_call'] = dataset['Day Charge'] / dataset['Day Calls']
```

### Remover features

```python
# aqui retiramos as colunas que nao queremos. 
# tiramos o churn pq Ã© a coluna q queremos prever
# churn Ã© separado, vai para y
x = dataset.drop(["Churn","Area Code", "Phone", "State", 
                  "VMail Message", "Day Mins", "Eve Mins",
                  "Night Mins", "Intl Mins", "Night Calls",
                  "Eve Calls", "Account Length"], axis = 1)
y = dataset["Churn"]
```

```python
x
```

```python
#substituir nulls por -1
x = x.fillna(-1)
```

### treino e teste modelo

```python
# importar o treino e o teste para o modelo
from sklearn.model_selection import train_test_split
```

```python
# repartiÃ§ao dos dados
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=26)
```

### Algoritmos de ML

#### Arvore de decisao

```python
# importar arvore de decisao
from sklearn.tree import DecisionTreeClassifier
```

#### Regressao logistica

```python
# importar a regressao logistica
from sklearn.linear_model import LogisticRegression
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

### CriaÃ§ao do Modelo

```python
# crio o modelo --- estÃ¡ vazio
modelo = DecisionTreeClassifier(random_state = 26)
```

```python
# crio o modelo --- estÃ¡ vazio
modelo = LogisticRegression(random_state = 0, max_iter =100000000000)
```

```python
modelo = RandomForestClassifier(random_state=0, n_estimators=100)
```

#### Treinar o modelo

```python
# treinar o modelo
# dar os dados ao modelo
modelo.fit(x_train, y_train)
```

#### testar o modelo

```python
# dou os dados de teste para prever
# as respostas do modelo vao para o y_pred
y_pred = modelo.predict(x_test)
```

```python
# o que preveu
y_pred
```

```python
# o que Ã© certo, real
y_test.values
```

## AvaliaÃ§ao do modelo

MÃ©tricas

#### Accuracy

```python
# importar a accuracy
from sklearn.metrics import accuracy_score\

# accuracy
accuracy_score(y_test, y_pred)
```

#### Matriz de confusao

```python
from sklearn.metrics import confusion_matrix

# matriz de confusao
confusion_matrix(y_test, y_pred)
```

#### Curva de ROC

```python
# curvas de roc
from sklearn.metrics import roc_auc_score

roc_auc_score(y_test, modelo.predict_proba(x_test)[:, 1])
```
