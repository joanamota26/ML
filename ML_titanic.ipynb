# Survived prediction



```python
import pandas as pd
```

```python
dataset = pd.read_csv("titanic_dataset.csv", sep = ",")
```

```python
dataset.head(5)
```

```python
# aqui retiramos as colunas que nao queremos. 
# tiramos o survived pq é a coluna q queremos prever
# survived é separado, vai para y
X = dataset.drop(["Survived", "Ticket", "Name", "Cabin", "PassengerId"], axis = 1)
y = dataset["Survived"]
```

```python
# passar strings para binario
X.loc[X["Sex"] == "male", "Sex_new"] = 1 
X.loc[X["Sex"] == "female", "Sex_new"] = 0
```

```python
X.loc[X["Embarked"] == "S", "Embarked_new"] = 0 
X.loc[X["Embarked"] == "C", "Embarked_new"] = 1
X.loc[X["Embarked"] == "Q", "Embarked_new"] = 2
```

```python
X
```

```python
X = X.drop(["Sex", "Embarked"], axis = 1)
```

```python
X = X.fillna(-1)
```

```python
# importar o treino e o teste para o modelo
from sklearn.model_selection import train_test_split
```

```python
#teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=26)
```

```python
# arvore de decisao
from sklearn.tree import DecisionTreeClassifier
```

```python
# crio o modelo --- está vazio
modelo = DecisionTreeClassifier(random_state = 26)
```

```python
# treinar o modelo
# dar os dados ao modelo
modelo.fit(X_train, y_train)
```

```python
# dou os dados de teste para prever
# as respostas do modelo vao para o y_pred
y_pred = modelo.predict(X_test)
```

```python
# o que preveu
y_pred
```

```python
# o que é certo, real
y_test.values
```

```python
# importar a accuracy
from sklearn.metrics import accuracy_score
```

```python
# ver a accuracy, testar o modelo
# ou seja, comparar o que preveu(y_pred) com as respostas certas reais (y_test)
accuracy_score(y_test, y_pred)
```

```python
# importar confusao de matriz
from sklearn.metrics import confusion_matrix
```

```python
# matriz de confusao
confusion_matrix(y_test, y_pred)
```
